```python
df1=sns.load_dataset('tips')
```

```python
df1
```

```python
sns.scatterplot(x=df1.total_bill,y=df1.tip)
```

```python
sns.relplot(x='total_bill',y='tip',data=df1) #relation plot similar to scatter plot
```

```python
df1
```

```python
df1['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='smoker') #hue parameter groups the data based on color,it can b categorical or numerical data
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='smoker') #style also performs group by operation based on the column and gvs different shape for each group
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='smoker',hue='smoker')
# both style n hue->same
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='smoker',hue='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,size='size')
```

```python
sns.relplot(x='total_bill',y='tip',kind='line',data=df1)
```

```python
# seperate graphs for each column based on the groups
sns.relplot(x='total_bill',y='tip',col='size',data=df1)
```

```python
# seperate graphs for each column based on the groups
sns.relplot(x='total_bill',y='tip',col='time',data=df1)
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,col='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,col='sex')
```

```python
# categorical plot
sns.catplot(x='day',y='tip',data=df1)
```

```python
sns.catplot(x='day',y='total_bill',data=df1)
```

```python
sns.catplot(x='day',y='tip',data=df1,kind='swarm') #not much useful plot
```

```python
sns.catplot(x='day',y='tip',data=df1,kind='box',hue='sex')
```
